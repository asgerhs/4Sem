```python
import random
import csv

class Course():
    def __init__(self, name, classroom, teacher, ETCS, grade=None):
        self.name = name
        self.classroom = classroom
        self.teacher = teacher
        self.ETCS = ETCS
        self.grade = grade
    
    def __repr__(self):
        return "{" + str(self.name) + ", " + str(self.classroom) + ", " + str(self.teacher) + ", " + str(self.ETCS) + ", " + str(self.grade) + "}"

class DataSheet():
    def __init__(self, courses):
        self.courses = courses
    
    def getGradesAsList(self):
        grades = []
        for course in courses:
            grades.append(course.grade)
        return grades
    
    def __repr__(self):
        return str(self.courses)

    


class Student():
    def __init__(self, name, gender, data_sheet, image_url):
        self.name = name
        self.gender = gender
        self.data_sheet = data_sheet
        self.image_url = image_url
    
    def getGradeAverage(self):
        return sum(self.data_sheet.getGradesAsList()) / len(self.data_sheet.getGradesAsList())
    
    def __repr__(self):
        return 'Student(%r, %r, %r, %r' % (self.name, self.gender, self.data_sheet, self.image_url)



names = ["Artem", "Asger", "Willi", "Andreas", "Martin", "Emil", "Henning", "Bo", "Filip", "Alex", "Elena", "Ryan", "Bob", "Anders", "Nikolaj"]
lastNames = ["Ivanov", "SÃ¸rensen", "Huusfeldt", "Vikke", "Frederiksen", "Svensmark"]
genders = ["Male", "Female"]
course1 = Course("Math", "103", "Pythagoras", 10, random.randint(0,13))
course2 = Course("Danish", "105", "H.C Andersen", 10, random.randint(0,13))
course3 = Course("English", "107", "Brokken Obama", 10, random.randint(0,13))
course4 = Course("Python", "107", "Hr. Snek", 20, random.randint(0, 13) )
courses = [course1, course2, course3]
scndCourse = [course1, course2, course3, course4]
datasheet = DataSheet(courses)
datasheet2 = DataSheet(scndCourse)
student1 = Student("Asger", "Male", datasheet, "fuck no lol")


def generateStudents(firstSem, secondSem):
    students = []
 
    while firstSem > 0:
        student = Student((names[random.randint(0,15)] + " " + lastNames[random.randint(0,5)]), genders[random.randint(0,1)],datasheet, "still no")
        firstSem = firstSem - 1
        students.append(student)
    
    while secondSem > 0:
        student = Student((names[random.randint(0,5)] + " " + lastNames[random.randint(0,5)]), genders[random.randint(0,1)],datasheet2, "still no")
        secondSem = secondSem - 1
        students.append(student)
    f = open("students.csv", "w")
    with f:
        studentInfo = ['stud_name', 'gender', 'course_name', 'teacher', 'ECTS', 'classroom', 'grade', 'img_url']
        writer = csv.DictWriter(f, fieldnames=studentInfo,quoting=csv.QUOTE_NONNUMERIC)
        writer.writeheader()
        for student in students:
            for courses in student.data_sheet.courses:
                writer.writerow({'stud_name' : student.name, 'gender' : student.gender, 'course_name' : courses.name, 'teacher' : courses.teacher, 'ECTS' : courses.ETCS, 'classroom' : courses.classroom, 'grade' : courses.grade, 'img_url' : student.image_url})
    
    return students

def readStudents():
    students = {}
    f = open('students.csv', 'r')

    with f: 
        reader = csv.DictReader(f)

        for row in reader:
            if row['stud_name'] not in students:
                students[row['stud_name']] = Student(row['stud_name'], row['gender'], DataSheet([Course(row['course_name'], row['classroom'], row['teacher'], row['ECTS'], row['grade'])]), row['img_url'])
            else:
                students[row['stud_name']].data_sheet.courses.append(Course(row['course_name'], row['classroom'], row['teacher'], row['ECTS'], row['grade']))

    for student_name in students:
        student = students[student_name]
        print('{}, {}, {}'.format(student.name, student.image_url, student.getGradeAverage()))






        
#datasheet.getGradesAsList()
#student1.data_sheet.getGradesAsList()
#student1.getGradeAverage()
students = generateStudents(7, 7)
#readStudents()
threeClosest(students)
```
