```python
from selenium import webdriver

base_url = "https://google.com"
browser = webdriver.Firefox()
browser.get(base_url)
browser.implicitly_wait(3)

search_field = browser.find_element_by_name("q")
search_field.send_keys("world population review")
search_field.submit()

results = browser.find_elements_by_xpath('//div[@class="r"]/a/h3')
results[0].click()

search_field = browser.find_element_by_name("query")
search_field.send_keys("Denmark")
search_field.submit()

search_result = browser.find_element_by_id("search_results")
result_table = search_result.find_element_by_id("content")
result = result_table.find_element_by_link_text("Denmark")
result.click()
```

```python
import bs4
import requests

r = requests.get('https://worldpopulationreview.com/countries/denmark-population/')
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')

print(soup.prettify())


```
